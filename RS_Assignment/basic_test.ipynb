## MovieLens 데이터 불러오기 및 전처리

```python
from calldata import call_dataset
from calldata import preprocessing
from calldata import similarity
from calldata import basic_cf_algorithm

import numpy as np
import pandas as pd

Call_MovieLens=call_dataset.CallRSData("dataset/ml-100k/u.data")
Mat_MovieLens=Call_MovieLens.CallMovieLens()

Preprocessing_MovieLens=preprocessing.Preprocessing_RSData(Mat_MovieLens)
pre_MovieLens=Preprocessing_MovieLens.Preprocessing() #전처리 완료

Similarity_MovieLens=similarity.All_Similarity(pre_MovieLens) # Similarity Class 불러오기
Basic_CF_MovieLens=basic_cf_algorithm.Basic_CF_Algorithm(pre_MovieLens) # basic cf algorithm 불러오기
```

```python
pre_MovieLens.head() #밑으로 USER, 좌우(column)로 ITEM
```
