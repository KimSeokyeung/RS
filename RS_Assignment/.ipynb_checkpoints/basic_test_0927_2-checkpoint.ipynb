row : user_id

1.user 중 rating을 많이 한 상위 1000명을 추출

2.artist 중 rating이 많이 된 상위 1000개를 추출 

```python
df_LastFM=pd.DataFrame(Num_Mat_LastFM.T)

# 모든 값에 log를 취함
pre_LastFM=(np.log(df_LastFM[df_LastFM[:]!=0])+1)


##### artist 기준 전처리 #####
# rating된 값이 많은 순서대로 index값 추출
# 상위 1000개를 제외하고 artist_index에 저장
artist_index=pre_LastFM.count().sort_values(ascending=False).index[1000:]
# artist_index를 제외한 상위 1000개의 artist_id만 남겨놓음
pre_LastFM=pre_LastFM.drop(artist_index, axis=1)


##### user 기준 전처리 #####
# rating된 값이 많은 순서대로 index값 추출
# 상위 1000개를 제외하고 user_index에 저장
user_index=pre_LastFM.count(axis=1).sort_values(ascending=False).index[1000:]
# user_index를 제외한 상위 1000개의 user_id만 남겨놓음
pre_LastFM=pre_LastFM.drop(user_index, axis=0)


# index 순서대로 sorting
pre_LastFM=pre_LastFM.sort_index()
pre_LastFM=pre_LastFM.sort_index(axis=1)
# NaN값 -> 0 으로 고치기
pre_LastFM=pre_LastFM.fillna(0)
pre_LastFM
```

# Matrix Similarity 측정

## 1. Cosine Similarity

### NaN 포함

```python
def COS(a):
    
    NumUsers = np.size(a, axis=0) # row size
    Sim = np.full((NumUsers, NumUsers), 0.0) # Sim matrix 초기화
    
    for u in range(0, NumUsers): # row : u번째 사용자
        for v in range(u, NumUsers): # column : v번째 아티스트
            
            innerDot = np.dot(a[u, ], a[v, ]) #내적
            NormU = np.linalg.norm(a[u, ]) #user의 크기
            NormV = np.linalg.norm(a[v, ]) #artist의 크기
            
            if(NormU == 0 or NormV ==0): #분모가 0이 되는 경우, 0으로 값 삽입
                Sim[u,v] = 0
            else: 
                Sim[u,v] = innerDot/(NormU * NormV) #COS 계산
                
            Sim[v,u] = Sim[u,v]
            
    return Sim
```

### NaN 제외

```python
def COS_except(a):
    
    NumUsers = np.size(a, axis=0) # row size
    Sim = np.full((NumUsers, NumUsers), 0.0) # Sim matrix 초기화
    
    for u in range(0, NumUsers): # row : u번째 사용자
        for v in range(u, NumUsers): # column : v번째 아티스트
            
            arridx_u = np.where(a[u, ] == 0)
            arridx_v = np.where(a[v, ] == 0) # NaN값 찾기
            arridx = np.concatenate((arridx_u, arridx_v), axis = None)
            
            U = np.delete(a[u, ], arridx)
            V = np.delete(a[v, ], arridx) # NaN값 제외
            
            innerDot = np.dot(U, V) #내적
            NormU = np.linalg.norm(U) #user의 크기
            NormV = np.linalg.norm(V) #artist의 크기
            
            if(NormU == 0 or NormV == 0): #분모가 0이 되는 경우, 0으로 값 삽입
                Sim[u,v] = 0
            else: 
                Sim[u,v] = innerDot/(NormU * NormV) #COS 계산
                
            Sim[v,u] = Sim[u,v]
            
    return Sim
```

## 2. PCC (Pearson Correlation Coefficient)

### PCC

```python
def PCC(a):
    
    NumUsers = np.size(a, axis=0) # row size
    Sim = np.full((NumUsers, NumUsers), 0.0) # Sim matrix 초기화
    
    mean = np.nanmean(np.where(a!=0, a, np.nan), axis=1) #두 사용자 모두 NaN이 아닌 것의 평균
    mean[np.isnan(mean)] = 0
    
    for u in range(0, NumUsers): # row : u번째 사용자
        for v in range(u, NumUsers): # column : v번째 아티스트
            
            arridx_u = np.where(a[u, ] == 0)
            arridx_v = np.where(a[v, ] == 0) # NaN값 찾기
            arridx = np.concatenate((arridx_u, arridx_v), axis = None) #join
               
            U = np.delete(a[u, ], arridx)
            V = np.delete(a[v, ], arridx) # NaN값 제외
            
            U = U - mean[u]
            V = V - mean[v]
            
            # COS과 동일
            innerDot = np.dot(U, V) #내적
            NormU = np.linalg.norm(U) # user의 크기
            NormV = np.linalg.norm(V) # artist의 크기
            
            if(NormU * NormV == 0): #분모가 0이 되는 경우, 0으로 값 삽입
                Sim[u,v] = 0
            else: 
                Sim[u,v] = innerDot/(NormU * NormV) #COS 계산
                
            Sim[v,u] = Sim[u,v]
            
    return Sim
```

```python
PCC(np.array(pre_LastFM))
```

### Constrained PCC

```python
def CPCC(a):
    
    NumUsers = np.size(a, axis=0) # row size
    Sim = np.full((NumUsers, NumUsers), 0.0) # Sim matrix 초기화
    
    median = np.nanmedian(np.where(a!=0, a, np.nan), axis=1) #두 사용자 모두 NaN이 아닌 것의 평균
 
    for u in range(0, NumUsers): # row : u번째 사용자
        for v in range(u, NumUsers): # column : v번째 아티스트
            
            arridx_u = np.where(a[u, ] == 0)
            arridx_v = np.where(a[v, ] == 0) # NaN값 찾기
            arridx = np.concatenate((arridx_u, arridx_v), axis = None) #join
               
            U = np.delete(a[u, ], arridx)
            V = np.delete(a[v, ], arridx) # NaN값 제외
            
            U = U - median[u]
            V = V - median[v]
            

            # COS과 동일
            innerDot = np.dot(U, V) #내적
            NormU = np.linalg.norm(U) #user의 크기
            NormV = np.linalg.norm(V) #artist의 크기
            
            Sim[u,v] = innerDot/(NormU * NormV) #COS 계산
            Sim[v,u] = Sim[u,v]
            Sim[np.isnan(Sim)] = -1 #NaN값 제거
            
    return Sim
```

## 3. Jaccard Measure

교집합 / 합집합

값이 1에 근접할 수록 좋음

교집합 = 둘 다 rating 한 값이 있는 것

합집합 = 한 명이라도 rating 한 값이 있는 것

LastFM 데이터에 0이 많이 포함되어 있다. 따라서 Jaccard Measure가 가장 적합한 방법이라 생각하여 이를 선택했다.

```python
def JAC(a):
    
    NumUsers = np.size(a, axis=0) # row size
    Sim = np.full((NumUsers, NumUsers), 0.0) # Sim matrix 초기화
    
    for u in range(0, NumUsers): # row : u번째 사용자
        for v in range(u, NumUsers): # column : v번째 아티스트
            
            U = np.array(a[u, ] > 0, dtype = np.int)
            V = np.array(a[v, ] > 0, dtype = np.int) # rating binary
            
            SumUV = U + V 
            #합 = 2 : 둘 다 rating / 합 = 1 : 둘 중 하나만 rating / 합 = 0 : 둘 다 rating X
            
            Inter = np.sum(np.array(SumUV > 1, dtype = np.int)) #교집합
            Union = np.sum(np.array(SumUV > 0, dtype = np.int)) #합집합
            
            if(Union == 0): #분모가 0이 되는 경우, 0으로 값 삽입
                tmp=0
            else:    
                tmp = Inter / Union # 교집합/합집합
            Sim[u,v] = tmp
            Sim[v,u] = Sim[u,v]
            
    return Sim
```

```python
JAC(np.array(pre_LastFM))
```

# 이외의 방법 

## Euclidean Distance

```python
def EUC(a):
    
    NumUsers = np.size(a, axis=0) # = row size
    Sim = np.full((NumUsers, NumUsers), 0.0) #Sim matrix 초기화
    
    for u in range(0, NumUsers): # u번째 사용자
        for v in range(u, NumUsers): # v번째 아티스트
            
            tmp = np.sum(np.square(a[u,]-a[v,])) #EUC값 계산
            
            Sim[u,v] = np.sqrt(tmp)
            Sim[v,u] = Sim[u,v]
            
    return Sim
```

```python
EUC(np.array(pre_LastFM))
```

## Mean Squared Difference

```python
def MSD(a):
    
    NumUsers = np.size(a, axis=0) # row size
    Sim = np.full((NumUsers, NumUsers), 0.0) #sim matrix 초기화
    
    for u in range(0, NumUsers): # u번째 사용자
        for v in range(u, NumUsers): #v번째 아티스트
            
            U = np.where(a[u, ] == 0, np.nan, a[u, ])
            V = np.where(a[v, ] == 0, np.nan, a[v, ]) # 0을 NaN으로 
            
            SquaredSum = np.square(U - V) #MSD값 계산
            SquaredSum = SquaredSum[~np.isnan(SquaredSum)] # Nan값 제거
            
            AllItems = np.size(SquaredSum, axis = 0) #NaN값을 제외한 크기
            
            if(AllItems == 0):
                tmp = 0
            else:
                tmp = np.sum(SquaredSum)/AllItems #Nan값을 제외한 합
                
            Sim[u,v] = tmp
            Sim[v,u] = Sim[u,v]
            
    return Sim
```

```python
MSD(np.array(pre_LastFM))
```

## JMSD

JAC + MSD

MSD값을 값이 클수록 좋게 바꿈

-> 결론 : 값이 클수록 좋은 것

```python
def JMSD (a, max): 
    return JAC(a) * (1-(MSD(a/max)))
```

```python
JMSD(np.array(pre_LastFM), max(pre_LastFM))
```
